## Converting to Date Time

```python
df_comp["date"] = pd.to_datetime(df_comp["date"], dayfirst=True)
```

```python
df_comp.info()
```

```python
df_comp.head()
```

```python
df_comp["date"].describe()
```

```python
df_comp.set_index("date", inplace=True)
```

```python
df_comp.head()
```

## Set Frequency

```python
df_comp = df_comp.asfreq('d')
```

```python
df_comp.head()
```

```python
# parameter b is for only business days
df_comp = df_comp.asfreq('b')
```

```python
df_comp.head(10)
```

```python
df_comp.isnull().sum()
```

```python
#need tofill missing values with fillna
#front filling will fill value based on previous day
#back filling will do opposite, and assign value from day after missing day

df_comp.fillna(method='ffill', inplace = True)
```

```python
df_comp.isna().sum()
```

## Clean up data - only want to view s&p 500

```python
df_comp.head()
```

```python
df_comp["Market Value"] = df_comp["spx"]
```

```python

df_comp.head()
```

```python
list(df_comp.columns)[0:-1]
```

```python
spDF = df_comp.drop(labels = list(df_comp.columns)[0:-1], axis = 1)
```

```python
spDF
```

## Splitting Data for training and test

```python
#Training will be historical and in order, not shuffled

size = len(spDF)

training = spDF.iloc[0:int(size*.8)]
test = spDF.iloc[int(size*.8):]
```
