## Random Walk

```python
randomWalkData = pd.read_csv("RandWalk.csv")
```

```python
randomWalkData["date"] =pd.to_datetime(randomWalkData["date"], dayfirst=True)
randomWalkData.set_index("date", inplace = True)
```

```python
randomWalkData = randomWalkData.asfreq('b')
```

```python
randomWalkData.head()
```

```python
fig, ax = plt.subplots(figsize = (20,14))
ax.plot(randomWalkData.index, randomWalkData["price"], label = 'Random Walk')
ax.plot(training.index, training["Market Value"], label = "S&P")
ax.legend()
```

## Stationarity

```python
import statsmodels.tsa.stattools as sts
```

```python
sts.adfuller(training["Market Value"])
```

```python
sts.adfuller(randomWalkData["price"].dropna())
```

## Seasonality
seasonality is cyclical patterns

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
s_dec_additive = seasonal_decompose(training["Market Value"], model="additive")

s_dec_additive.plot()
```

* we see spikes in residual at the 2 points where the markets crashed in .com era and housing market

```python
s_dec_multiplicative = seasonal_decompose(training["Market Value"], model="multiplicative")

s_dec_multiplicative.plot()
```

## ACF

```python
import statsmodels.graphics.tsaplots as sgt
```

```python
#lags to 40 means the last 40 lags or periods before the current one

sgt.plot_acf(training["Market Value"], lags = 40, zero=False, title="ACF")
```

```python
#ACF for randomwalk
sgt.plot_acf(randomWalkData["price"], zero = False, lags = 40, title = "Random Walk ACF")
```

## PACF

```python

sgt.plot_pacf(training["Market Value"], zero = False, lags = 40, title = "PACF S&P", method=("ols"))
```

```python
sgt.plot_pacf(randomWalkData["price"], zero = False, lags = 40, title = "PACF RandomWalk", method=("ols"))
```
