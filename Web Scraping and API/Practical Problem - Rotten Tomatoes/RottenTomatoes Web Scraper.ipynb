# Read in Libaries and Get Page Source Code

```python
import requests
from bs4 import BeautifulSoup
import urllib3
```

```python
base_url = 'https://editorial.rottentomatoes.com/guide/140-essential-action-movies-to-watch-now/'
```

```python
requests.get(base_url)
```

```python
htmlCode = requests.get(base_url).content
```

```python
soup = BeautifulSoup(htmlCode)
```

# Obtain Parent Div for Movie Reviews

```python
movieReviews = soup.find_all('div', class_ = "row countdown-item")
```

```python
movieReviews
```

# Test element for finding content desired

### Finding Title of Movies

```python
testDiv = movieReviews[0]
```

```python
testHeading = testDiv.find('h2')
```

```python
testHeading
```

```python
testHeading.find('a').text
```

### Finding Year of Movies

```python
int(testHeading.find('span', class_ = 'subtle start-year').text.strip('()'))
```

### Finding Score of Movie

```python
testHeading.find('span', class_ = 'tMeterScore').text
```

### Applying List Comprehension

```python
titles = [div.find('h2').find('a').text for div in movieReviews ]
```

```python
year =  [int(div.find('h2').find('span', class_ = 'subtle start-year').text.strip('()')) for div in movieReviews]
```

```python
scores = [int(div.find('h2').find('span', {"class": "tMeterScore"}).text.strip('%')) for div in movieReviews]
```

```python
list(zip(titles, year, scores))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data = list(zip(titles, year, scores)), columns=["Title", "Year", "Scores"])
```

```python
df
```
